```python
import pandas as pd
import json
```

```python
# 1. Load the raw JSON export from Label Studio
with open('../data/annotated/sentiment-reviews-label-2026-03-09-09-51.json', 'r', encoding='utf-8') as f:
    raw_annotations = json.load(f)

# 2. Extract the cleaned text and the sentiment labels
# We use a list comprehension for efficiency
processed_data = []

for task in raw_annotations:
    try:
        # Accessing the text you already cleaned in your previous script
        text = task['data']['cleaned_content']
        
        # Accessing the first annotation result
        # Label Studio nested path: annotations -> result -> value -> choices
        label = task['annotations'][0]['result'][0]['value']['choices'][0]
        
        processed_data.append({
            "cleaned_review": text,
            "sentiment": label
        })
    except (IndexError, KeyError) as e:
        # This catches any skipped tasks or malformed entries
        print(f"Skipping task ID {task.get('id')}: No label found.")
```

```python
# 3. Create the 'Golden Dataset'
df_labeled = pd.DataFrame(processed_data)

# 4. Save to a clean CSV for the next stage (EDA)
df_labeled.to_csv('final_labeled_fintech_data.csv', index=False)

print(f"Successfully merged {len(df_labeled)} records.")
df_labeled.head()
```

```python
# Check 1: Ensure no labels were missed
missing_labels = df_labeled['sentiment'].isnull().sum()

# Check 2: Verify the unique labels (Check for typos like 'Positve' vs 'Positive')
unique_labels = df_labeled['sentiment'].unique()

# Check 3: Check for empty text strings
empty_text = (df_labeled['cleaned_review'].str.strip() == "").sum()

print(f"Missing Labels: {missing_labels}")
print(f"Unique Sentiments: {unique_labels}")
print(f"Empty Text Records: {empty_text}")
```
